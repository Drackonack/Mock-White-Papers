### **Requirements**

```python
from sklearn.metrics import accuracy_score
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG16, efficientnet, resnet50

import itertools
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
```

### **Dataset download, Priming, and visualisation**

```python
cifar10 = tf.keras.datasets.cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print("Training data shape:", x_train.shape)
print("Training labels shape:", y_train.shape)
print("Testing data shape:", x_test.shape)
print("Testing labels shape:", y_test.shape)
```

```python
y_train = y_train.flatten()
y_test = y_test.flatten()

classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
df = pd.DataFrame({"Class": [classes[label] for label in y_train]})

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="Class")
plt.title("Class Distribution in CIFAR-10 Training Set")
plt.xticks(rotation=55)
plt.show()
```

```python
input_shape = (32, 32, 3)

x_train=x_train.reshape(x_train.shape[0], x_train.shape[1], x_train.shape[2], 3)
x_train=x_train / 255.0
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2], 3)
x_test=x_test / 255.0
```

```python
y_train = tf.one_hot(y_train.astype(np.int32), depth=10)
y_test = tf.one_hot(y_test.astype(np.int32), depth=10)
```

<hr>

### **Helper Functions**

```python
def confusion_matrix(model, x_test, y_test):
    """
    Plots a confusion matrix for a given model, test data, and test labels.

    Parameters:
    model (tf.keras.Model): The trained model.
    x_test (numpy.ndarray): The test data.
    y_test (numpy.ndarray): The test labels.

    """
    y_pred = model.predict(x_test)
    y_pred = np.argmax(y_pred, axis=1)
    y_test = np.argmax(y_test, axis=1)
    cm = tf.math.confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix")
    plt.show()
```

```python
def plot_performance(history):
    """
    Plots the training loss and accuracy over epochs from a history object.

    Parameters:
    history (tf.keras.callbacks.History): The history object returned from model.fit().
    """
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss', color='blue')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training Loss over Epochs')
    plt.legend()
    plt.grid()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy', color='orange')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training Accuracy over Epochs')
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()
```

```python
def test_error(model, x_test, y_test):
    """
    Calculate the test error rate and accuracy of a trained model.

    Parameters:
    model (tf.keras.Model): The trained model.
    x_test (numpy.ndarray): The test data.
    y_test (numpy.ndarray): The true labels for the test data (one-hot encoded).

    Returns:
    float: Test error rate.
    float: Accuracy.
    """
    y_pred_prob = model.predict(x_test)

    y_pred = np.argmax(y_pred_prob, axis=1)

    y_test_labels = np.argmax(y_test, axis=1)

    accuracy = accuracy_score(y_test_labels, y_pred)
    error_rate = 1 - accuracy

    print(f'Test Accuracy: {round(accuracy, 4) * 100}%')
    print(f'Test Error Rate: {round(error_rate, 4) * 100}%')
```

### **Initial Model**

```python
batch_size = 32
num_classes = 10
```

```python
def student_model(input_shape, num_classes):
    model = models.Sequential()

    model.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))

    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D((2, 2)))

    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))

    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D((2, 2)))

    model.add(layers.Conv2D(256, (3, 3), activation='relu'))
    model.add(layers.Conv2D(256, (3, 3), activation='relu'))

    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D((2, 2), padding='same'))

    model.add(layers.Flatten())

    model.add(layers.Dense(4096, activation='relu'))
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(4096, activation='relu'))

    model.add(layers.Dense(num_classes, activation='softmax'))

    return model
```

```python
proposed_model = student_model(input_shape, num_classes)
proposed_model.summary()
```

```python
proposed_model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
proposed_model_history = proposed_model.fit(x_train, y_train, batch_size=batch_size, epochs=10)
```

```python
proposed_model_accuracy = proposed_model_history.history['accuracy'][-1]
print(f"Final Epoch Training Accuracy: {round(proposed_model_accuracy, 4)*100}%")
test_error(proposed_model, x_test, y_test)
```

```python
confusion_matrix(proposed_model, x_test, y_test)
```

```python
plot_performance(proposed_model_history)
```

### **ResNet50 Fine Tuning**

```python
def resnet_transfer_model(input_shape, num_classes):
    base_model = resnet50.ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)

    top_model = base_model.output
    top_model = layers.Flatten()(top_model)
    top_model = layers.Dense(4096, activation='relu')(top_model)
    top_model = layers.Dropout(0.3)(top_model)
    top_model = layers.Dense(4096, activation='relu')(top_model)
    top_model = layers.Dense(num_classes, activation='softmax')(top_model)

    model = models.Model(inputs=base_model.input, outputs=top_model)

    return model
```

```python
fine_tuning_model_resnet = resnet_transfer_model(input_shape, num_classes)
fine_tuning_model_resnet.summary()
```

```python
fine_tuning_model_resnet.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = fine_tuning_model_resnet.fit(x_train, y_train, batch_size=batch_size, epochs=10)
```

```python
final_training_accuracy = history.history['accuracy'][-1]
print(f"Final Epoch Training Accuracy: {round(final_training_accuracy, 4)*100}%")
test_error(fine_tuning_model_resnet, x_test, y_test)
```

```python
confusion_matrix(fine_tuning_model_resnet, x_test, y_test)
```

```python
plot_performance(history)
```

### **Improved Model**

```python
def improved_student_model(input_shape, num_classes):
    inputs = tf.keras.Input(shape=input_shape)

    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2, 2))(x)

    residual = layers.Conv2D(128, (1, 1), padding='same')(x)
    x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Add()([x, residual])
    x = layers.MaxPooling2D((2, 2))(x)

    residual = layers.Conv2D(256, (1, 1), padding='same')(x)
    x = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Add()([x, residual])
    x = layers.MaxPooling2D((2, 2))(x)

    residual = layers.Conv2D(512, (1, 1), padding='same')(x)
    x = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(x)
    x = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Add()([x, residual])
    x = layers.MaxPooling2D((2, 2))(x)

    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dense(1024, activation='relu')(x)
    x = layers.Dropout(0.4)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs, outputs)
    return model

input_shape = (32, 32, 3)
num_classes = 10
```

Key improvements as inspired by resnet, vgg and efficientnet.
more convolutional layers to be similar to VGG
residual netowrks included to be more similar to ResNet
Gloabal Average pooling instead of dense layers at the end reducing parameters but maintaining performance like efficientnet

```python
improved_cnn = improved_student_model(input_shape, num_classes)
improved_cnn.summary()
```

```python
improved_cnn.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```
